# Chain Rule Walkthrough in Code
We compute gradients for a simple computation graph and compare manual math with PyTorch autograd.

## Example
Let:

* u = x * y
* v = u + x
* z = v ** 2

We want dz/dx and dz/dy.

```python
import torch

x = torch.tensor(2.0, requires_grad=True)
y = torch.tensor(3.0, requires_grad=True)

u = x * y
v = u + x
z = v ** 2

z.backward()
x.grad, y.grad
```

Manual chain rule
- u = x * y
- v = u + x
- - z = v^2

- dz/dv = 2v
- dv/du = 1
- dv/dx = 1
- du/dx = y
- du/dy = x

dz/dx = dz/dv * (dv/du * du/dx + dv/dx)

dz/dy = dz/dv * dv/du * du/dy

The next cell confirms the numeric result.


```python
x_val, y_val = 2.0, 3.0
u_val = x_val * y_val
v_val = u_val + x_val
dz_dv = 2 * v_val
dz_dx = dz_dv * (1 * y_val + 1)
dz_dy = dz_dv * 1 * x_val

dz_dx, dz_dy
```
